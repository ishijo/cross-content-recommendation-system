```python
def getDescription(show_name,mode):
    show_name_l = show_name.lower()
    show_name_s = re.sub('[^a-zA-Z0-9_]',' ',show_name_l)
    show_name = re.sub(' +', ' ',show_name_s.strip())
    if mode.lower() in ['television', 'tv', 'tvshow', 'tv-show', 'tv - show', 'show', 'series']:
        try:
            url = 'https://www.rottentomatoes.com/tv/'+'-'.join(show_name.split(' '))
            agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
            response = requests.get(url, headers=agent)
            soup = BeautifulSoup(response.content, 'html.parser')
            soup_obj = soup.find('p',{'data-qa':'series-info-description'})
            text = soup_obj.text.strip()
        except:
            url = 'https://www.rottentomatoes.com/tv/'+'_'.join(show_name.split(' '))
            agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
            response = requests.get(url, headers=agent)
            soup = BeautifulSoup(response.content, 'html.parser')
            soup_obj = soup.find('p',{'data-qa':'series-info-description'})
            text = soup_obj.text.strip()     

    elif mode.lower() in ['movie', 'movies', 'film', 'films']:
        try:
            url = 'https://www.rottentomatoes.com/m/'+'-'.join(show_name.split(' '))
            agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
            response = requests.get(url, headers=agent)
            soup = BeautifulSoup(response.content, 'html.parser')
            soup_obj = soup.find('p',{'data-qa':'movie-info-synopsis'})
            text = soup_obj.text.strip()
        except:
            url = 'https://www.rottentomatoes.com/m/'+'_'.join(show_name.split(' '))
            agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
            response = requests.get(url, headers=agent)
            soup = BeautifulSoup(response.content, 'html.parser')
            soup_obj = soup.find('p',{'data-qa':'movie-info-synopsis'})
            text = soup_obj.text.strip()    
    return text
```
