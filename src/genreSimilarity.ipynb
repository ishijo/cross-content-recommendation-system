```python
import pandas as pd
import numpy as np
import pickle
import spacy
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
# file = open('imdb_genres_freq','ab')
# pickle.dump(sorted_genres_i,file)
# file.close()

file = open('goodreads_genres_freq','rb')
sorted_genres_g = pickle.load(file)
file.close()

# file = open('imdb_genres_freq','ab')
# pickle.dump(sorted_genres_i,file)
# file.close()

file = open('imdb_genres_freq','rb')
sorted_genres_i = pickle.load(file)
file.close()
```

```python
g_genres = list(sorted_genres_g.keys())
i_genres = list(sorted_genres_i.keys())
print(len(g_genres))
print(len(i_genres))
```

```python
print(g_genres)
```

```python
print(i_genres)
```

```python
sim_dict = {}
for imdb_genre in i_genres:
    for gr_genre in g_genres:

        imdb = nlp.vocab[imdb_genre]
        gr = nlp.vocab[gr_genre]

        k = imdb_genre + ' - ' + gr_genre
        sim_dict[k] = imdb.similarity(gr)
```

```python
sorted_sim_dict = dict(sorted(sim_dict.items(), key=lambda x: x[1], reverse=True))
sorted_sim_dict
```

# After Stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

def stemming_genres(genre_lis):
    ps = PorterStemmer()
    stem_genres = []
    for substr in genre_lis:
        words = substr.split(' ')
        stem_genres.append(" ".join([ps.stem(word) for word in words]))
    return stem_genres
    
```

```python
print(stemming_genres(g_genres))
```

```python
print(stemming_genres(i_genres))
```

```python
stem_g_genres = stemming_genres(g_genres)
stem_i_genres = stemming_genres(i_genres)
```

```python
sim_dict2 = {}
for imdb_genre in stem_i_genres:
    for gr_genre in stem_g_genres:

        imdb = nlp.vocab[imdb_genre]
        gr = nlp.vocab[gr_genre]

        k = imdb_genre + ' - ' + gr_genre
        sim_dict2[k] = imdb.similarity(gr)
```

```python
sorted_sim_dict2 = dict(sorted(sim_dict2.items(), key=lambda x: x[1], reverse=True))
sorted_sim_dict2
```
