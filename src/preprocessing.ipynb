```python
import pandas as pd
import numpy as np
import re
import os
from os.path import dirname
import missingno as msno
import pickle
```

```python
data_path = os.path.join(dirname(os.getcwd()),'data')

goodreads_data_path = os.path.join(data_path,'goodreads')
goodreads = pd.read_csv(goodreads_data_path+'/goodreads_data.csv',index_col=0)

imdb_data_path = os.path.join(data_path,'imdb_filtered')
imdb = pd.read_csv(imdb_data_path+'/IMDB_sample2.csv',index_col=0)
```

```python
display('Goodreads',goodreads.head(2),'IMDB',imdb.head(2))
```

```python
print(goodreads.dtypes, '\n\n',imdb.dtypes)
```

```python
imdb['runtimeMinutes'].value_counts().head()
```

```python
## Checking for non numeric values in the runtimeMinutes column.

for i in imdb['runtimeMinutes'].value_counts().index:
    if re.search('[^0-9]',i):
        print(i)
```

```python
imdb['runtimeMinutes'].replace('\\N',np.nan,inplace=True)
```

```python
goodreads['Num_Ratings'] = goodreads['Num_Ratings'].apply(lambda x: int(''.join(x.split(','))))
imdb[['runtimeMinutes']]  = imdb[['runtimeMinutes']].apply(pd.to_numeric)
```

```python
goodreads.drop('URL',axis=1,inplace=True)
imdb.drop(['types','attributes','ordering','endYear','language','originalTitle','title','isOriginalTitle','isAdult'],axis=1,inplace=True)
```

```python
msno.matrix(goodreads);
```

```python
msno.matrix(imdb);
```

```python
g = open('goodreads_preprocessed','ab')
pickle.dump(goodreads,g)
g.close()
```

```python
i = open('imdb_preprocessed','ab')
pickle.dump(imdb,i)
i.close()
```
